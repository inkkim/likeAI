 _1)뚜봇_
  * 2015년 '두드리소'를 거쳐 2017년 정식 출시된 대구시 인공지능 민원 상담사

  * 카카오톡과 연계하여 24시간 여권, 자동차등록, 시정일반, 지역축제 관련 민원 상담 가능

  * 기업의 인트라넷, 웹, 슬랙 등 원하는 서비스 전달채널을 선택하여 활용 가능

- __번역 (Translation)__
 
  _1) 구글 번역_

  * 2006년 구글에서 출시한 무료 번역 서비스로 초기 4개 국어(영어, 중국어, 프랑스어, 스페인어)에서 현재 약 100개 언어 지원

  * 2016년 '딥러닝(Deep Learning)'과 '인공신경망(Artificial Neural Network, ANN)' 기반의 번역 서비스 도입
  
    + 한국어를 포함한 8개 언어 조합에 자체 개발한 인공신경망 번역 기술인 '구글 신경망 기계번역(Google's Nerual Machine Translation, GNMT)' 적용

  * 그외 광학문자 인식, 손글씨 인식, 실시간 통역, 음성 번역 기능 제공

  _2) 네이버 파파고_

 * 2016년 네이버랩스에서 출시한 무료 통번역 서비스로 한국어, 중국어, 일본어, 영어 를 비롯하여 총 14개 언어 지원

 * 네이버 자체 개발한 인공신경망 번역기술(Naver Neural Machine Translation, N2MT) 적용
   + 구글 GNMT와 비교할 때, 번역기가 네이버의 방대한 한글 데이터를 학습(모델링)함으로써, 번역 품질이 향상되어 한국어의 높임말 번역도 가능

 * 그외 텍스트 번역, 음성, 대화, 이미지, 학습도구 기능으로 구성
   + 학습 도구인 '글로벌 회화'와 '키즈'는 여행 전 기본적인 회화를 익힐 때, 아이와 함께 영단어 암기를 할 때 유용

- __인공지능 비서 (ArtificiaI Intelligence Assistant)__

 _1)구글 어시스턴트_

 * 삼성 냉장고, TV, 웨어러블 기기 등 삼성 스마트 기기에서 호환성이 높다.

- __음성 합성 기술 (Voice Synthesis)__

 _1)네이버 클로바 보이스_
 * 네이버 클로바 스마트 스피커, 파파고 외국어 발음, 지도 음성 안내 등 서비스에서 음성으로 사용
 * 네이버 Speaker Adapation 기법을 사용하여 약 40분 수준의 음성 녹음으로 고품질 음성 합성기 제작 가능
  + 네이버에서 개발한 NES(Natural End-to-End Speech Synthesis)기술 적용
 * 실시간 음성 생성이 가능한 Neural Vocoder를 사용하여, 실제 사람의 음성에 가까운 합성 음성을 제공
 * 특정 분야에 필요한 목소리 특성을 살려 합성음을 만드는 기법인 HDTS(High-quality DNN Text-to-Speech)기술 적용
   + 클로바 스마트 스피커와 오디오북 유인나 목소리 제작
   + 뉴스, 식당, 라디오 등 상황에 맞는 음성톤 제작 가능
 
 _2)Selvy_
 * 셀바스 AI(국내기업)에서 제공하는 인공지능 기반 음성인식/합성 솔루션
 * Selvy STT(Speech to Text/Speech Recognition)는 소리 정보를 분석하여 문자, 명령어 및 다양한 형태의 정보로 변환해주는 솔루션
    + 음성 검출, 특징 추출, 음향 모델, 탐색 및 후 처리에 대한 전 과정의 핵심요소 독자 개발
    + 고립 명령어, 대용량 POI, 가변어, 연결어, 핵심어 인식 등 인식 단어 확장/변경 가능
    + 서버형 대용량 엔진, 임베디드 단말형 엔진
    + 한국어 특성과 발성 습관을 반영하여 한국어 분야 세계 최고 수준 인식률
 * Selvy TTS(Text to Speech/Speech Synthesis)는 문자 정보를 분석하여 자연스러운 사람 목소리로 바꾸어 말해줌
    + 서버형 솔루션을 통해 플랫폼 및 사양에 구애 받지 않는 대용량 고성능 TTS
    + 다양한 임베디드/모바일 플랫폼에 최적화
    + 한국어/영어/중국어/일본어/스페인어/프랑스어/포르투갈어 지원
    + 콜센터, 방송, 교육 및 단말 제조 분야 등 다양한 산업 분야에 적용
      - 남성/여성/아동/캐릭터와 같이 다양한 종류의 목소리로 합성가능
      - 대화체/낭독체 등 다양한 스타일에 맞는 화자 목소리 제공
